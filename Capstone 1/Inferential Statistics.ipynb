###  Manhattan and Brooklyn are the boroughs of NYC with most number of Airbnb listings. We want to evaluate if the prices of Airbnb lisitngs are significantly different in the two boroughs.
<br><br>
* **H0 (Null Hypothesis) --> Mean price of Manhattan listings = Mean price of Brooklyn listings**
* **H1 (Alternate Hypothesis) --> Mean price of Manhattan listings != Mean price of Brooklyn listings**

```python
manhattan_price = df[df['neighbourhood_group_cleansed']=='Manhattan']['price'].values
brooklyn_price=df[df['neighbourhood_group_cleansed']=='Brooklyn']['price'].values
all_listings = df['price'].values

print('Mean price of Manhattan listings is : ', np.mean(manhattan_price))
print('Mean price of Brooklyn listings is : ', np.mean(brooklyn_price))
print('Mean price of all listings is : ', np.mean(all_listings))
```

```python
t, p_val = ttest_ind(manhattan_price, brooklyn_price)

alpha=0.05
print('Null Hypothesis : Mean price of Manhattan listings = Mean price of Brooklyn listings \np-value = ', p_val)
if p_val < alpha:  
    print("Verdict: The null hypothesis can be rejected.")
else:
    print("Verdict: The null hypothesis cannot be rejected.")
```

* **The hypothesis testing suggests that the null hypothesis can be rejected, indicating that we have enough evidence to conclude that the mean price of Manhattan listings are statistically different from mean price of Brooklyn listings. Manhattan Airbnbs seem to be in general pricier than Brooklyn ones.**

### Does host response time have an effect on the review communication scores? We will divide the data into 2 samples - one where host replied within hours and another where the host took longer to respond.
<br><br>
* **H0 (Null Hypothesis) --> Mean communication score when host responds within hours - Mean communication score when host takes longer = 0**
* **H1 (Alternate Hypothesis) --> Mean communication score when host responds within hours - Mean communication score when host takes longer > 0**

```python
all_listings= df['review_scores_communication'].values
less_time = df[df['host_response_time'].str.contains('hour')]['review_scores_communication'].values
more_time = df[~df['host_response_time'].str.contains('hour')]['review_scores_communication'].values

print('Mean communication score when host responds within hours : ', np.mean(less_time))
print('Mean communication score when host takes longer to respond : ', np.mean(more_time))

less_time_shifted=less_time-np.mean(less_time)+np.mean(all_listings)
more_time_shifted=more_time-np.mean(more_time)+np.mean(all_listings)
```

```python
def bootstrap_replicate_1d(data, func):
    """Draw one bootstrap replicate."""
    return func(np.random.choice(data, size=len(data)))
    

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
# Compute 10,000 bootstrap replicates from shifted arrays
bs_reps_less_time = draw_bs_reps(less_time_shifted, np.mean, 10000)
bs_reps_more_time = draw_bs_reps(more_time_shifted, np.mean, 10000)

# Get replicates of difference of means: bs_replicates
bs_reps = bs_reps_less_time-bs_reps_more_time

# Compute and print p-value: p
p_val = np.sum(bs_reps>0)/len(bs_reps)

print('Null Hypothesis : Mean communication score when host responds within hours - Mean communication score when host takes longer = 0 \np-value = ', p_val)
if p_val < alpha:  
    print("Verdict: The null hypothesis can be rejected.")
else:
    print("Verdict: The null hypothesis cannot be rejected.")
    
```

* **Based on the bootstrapping hypothesis test, we can conclude there the time taken by the host to respond does not have an effect on the review communication scores.**
